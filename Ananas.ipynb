```python
# 1. On crée la "recette" pour formater en NOM Prénom
def formater_nom_prenom(texte):
    # Si la case est vide ou contient 'Nan', on la laisse vide
    if pd.isna(texte) or str(texte).strip() == '' or str(texte).strip() == 'Nan':
        return ''
    
    # On découpe le texte en mots
    mots = str(texte).split()
    
    # S'il n'y a qu'un seul mot, on le met en majuscules
    if len(mots) == 1:
        return mots[0].upper()
    
    # S'il y a plusieurs mots : le 1er en MAJUSCULES, le reste avec une Majuscule au début
    nom = mots[0].upper()
    prenom = " ".join(mots[1:]).title()
    
    return f"{nom} {prenom}"

# 2. On liste tes colonnes de noms
colonnes_noms = ['nom_enqueteur', 'nom_repondant', 'nom_resp']

# 3. On applique la recette à tes colonnes
for col in colonnes_noms:
    if col in df.columns: # Sécurité pour vérifier que la colonne existe
        df[col] = df[col].apply(formater_nom_prenom)

# On affiche un petit aperçu pour vérifier !
print(df[['nom_enqueteur', 'nom_repondant']].head())
```
